```python
#scatter plot for  budget vs gross 
plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Scatterplot for budget vs gross earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for Film')
plt.show()
```

```python
#regression plot for 
sns.regplot(x='budget',y='gross',data=df,scatter_kws={'color':'red'},line_kws={'color':'blue'})
```

```python
#the above plot shows us a positive correlation but we dont know how much are they correlated
#so will use corr to see the correlation between numeric elements

#so inorder to do that without any errors we will first makesure we include only numeric values by using the code below
numeric_df=df.select_dtypes(include=['int64','float64'])

#this below code will then show us the correlation matrix
correlation_matrix=numeric_df.corr() #pearson,kendle,spearman
print(correlation_matrix)
```

```python
# the above correlation matrix shows us high correlation btwn gross and budget
```

```python
#now we will be plotting the correlation matrix to clearly show the correlation between the numeric elements
correlation_matrix = numeric_df.corr()

sns.heatmap(correlation_matrix, annot=True)#annot=true displays the correlation values on the heatmap
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```
